# Introduction
**The research question:** Is human freedom the same across space and time? This research question invites us to assess the degree to which people around the world enjoy human freedom as defined by key personal, civil and economic liberties.

The research question also provides an opportunity to evaluate the relationship among some of these civil and econmic liberties in order to evaluate their level of mutual association in relation to human freedom. Checking for these interdependences would allow us to then draw inferencial statistics and perhaps build a robust predictive model for forecasting Human Freedom and anticipate, mitigate or prevent potential threats to it.

**Rationale for working on the topic:** My interest in exploring this topic comes from the fact that I work for a law enforcement agency and we’re often called to restraint individuals’ exercise of their freedom in the name of public safety or for national security. Understanding the impact of our actions on our fellow citizens and how far not to go with the restrictions to liberties are paramount to good public service.

**The data:** For this assignment, I will use the Human Freedom Index which presents the state of human freedom worldwide based on personal, civil, and economic freedom indicators. Our selected working data cover the period 2019 - 2021.
The Human Freedom index uses 86 distinct indicators of personal and economic freedom and places each country on a scale of 0 to 10, where a score of 10 represents the most freedom, for personal freedom and economic freedom. This analysis will focus on just a few areas , including, personal freedom, economic freedom and the rule of law.

*Source:*  The Human Freedom Index 2023
*Link:* https://www.cato.org/human-freedom-index/2023

*Citation:* Ian Vázquez, Fred McMahon, Ryan Murphy, and Guillermina Sutter Schneider, The Human Freedom Index 2023: A Global Measurement of Personal, Civil, and Economic Freedom (Washington: Cato Institute and the Fraser Institute, 2023).
