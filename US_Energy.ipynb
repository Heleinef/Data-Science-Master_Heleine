```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data preparation
data = {
    'State': ['MS', 'TX', 'AR', 'LA', 'OK', 'KY', 'WV', 'MI', 'NM', 'OH'],
    'Food Insecurity (%)': [16.2, 16.9, 18.9, 16.2, 15.4, 14.5, 13.7, 13.0, 12.9, 12.8],
    'Very Low Food Security (%)': [6.1, 6.4, 6.7, 6.1, 5.7, 6.5, 4.1, 5.3, 5.0, 5.3]
}
df = pd.DataFrame(data)
df_long = pd.melt(df, id_vars='State', var_name='Category', value_name='Percentage')

# Create pivot table for heatmap
heat_df = df_long.pivot(index='Category', columns='State', values='Percentage')

# Plot heatmap
plt.figure(figsize=(12, 4))
sns.heatmap(heat_df, annot=True, cmap="Reds", fmt=".1f", linewidths=.5, cbar_kws={'label': 'Percentage (%)'})
plt.title("Heatmap of Food Insecurity by State")
plt.xlabel("State")
plt.ylabel("Category")
plt.tight_layout()

# Save the heatmap
plt.savefig("food_insecurity_heatmap.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/Heleinef/Data-Science-Master_Heleine/main/foodsecurity_hh_w_children_2023.csv"
df = pd.read_csv(url)

# Clean column names
df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace("-", "_").str.replace(",", "")

# Filter for national-level data
df_national = df[df['Subcategory'].str.lower() == 'all households'].copy()

# Convert columns to numeric
df_national["Year"] = pd.to_numeric(df_national["Year"], errors='coerce')
df_national["Food_insecure_households_percent"] = pd.to_numeric(df_national["Food_insecure_households_percent"], errors='coerce')
df_national["Households_with_very_low_food_security_among_children_percent"] = pd.to_numeric(
    df_national["Households_with_very_low_food_security_among_children_percent"], errors='coerce')

# Drop missing values
df_national = df_national.dropna(subset=[
    "Year",
    "Food_insecure_households_percent",
    "Households_with_very_low_food_security_among_children_percent"
])

# Prepare long format for Seaborn
df_long = pd.melt(
    df_national,
    id_vars="Year",
    value_vars=[
        "Food_insecure_households_percent",
        "Households_with_very_low_food_security_among_children_percent"
    ],
    var_name="Category",
    value_name="Percent"
)

# Make category names more readable
category_labels = {
    "Food_insecure_households_percent": "Food Insecure Households",
    "Households_with_very_low_food_security_among_children_percent": "Very Low Food Security (Children)"
}
df_long["Category"] = df_long["Category"].map(category_labels)

# Plot
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

plot = sns.lineplot(
    data=df_long,
    x="Year",
    y="Percent",
    hue="Category",
    marker="o",
    linewidth=2.5
)

plt.title("Trends in Food Insecurity Among U.S. Households with Children (20012023)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Percent of Households")
plt.legend(title="Food Security Category", loc="upper right")
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean the data
url = "https://raw.githubusercontent.com/Heleinef/Data-Science-Master_Heleine/main/foodsecurity_hh_w_children_2023.csv"
df = pd.read_csv(url)

# Drop rows with missing values in key columns
df_clean = df.dropna(subset=[
    'Year',
    'Category',
    'Subcategory',
    'Households with food-insecure children-percent'
])

# Convert columns to appropriate types
df_clean['Year'] = pd.to_numeric(df_clean['Year'], errors='coerce')
df_clean['Percent'] = pd.to_numeric(df_clean['Households with food-insecure children-percent'], errors='coerce')

# Filter to only Household Composition and Census Geographic Region
df_filtered = df_clean[
    df_clean['Category'].str.lower().isin(['household composition', 'census geographic region'])
].copy()

# Clean Category labels for display
df_filtered['Category'] = df_filtered['Category'].str.title()

# Set theme and create FacetGrid
sns.set_theme(style="whitegrid")
g = sns.FacetGrid(
    df_filtered,
    col="Category",
    hue="Subcategory",
    sharey=True,
    height=5,
    aspect=1.4,
    col_order=["Household Composition", "Census Geographic Region"]
)

# Draw line plots
g.map_dataframe(
    sns.lineplot,
    x="Year",
    y="Percent",
    marker="o"
)

# Customize titles, legends, and layout
g.add_legend(title="Subcategory", bbox_to_anchor=(1.05, 1), loc="upper left")
g.set_axis_labels("Year", "% with Food-Insecure Children")
g.set_titles("{col_name}")
plt.subplots_adjust(top=0.85)
g.fig.suptitle("Trends in Food Insecurity Among Children by Household Type and Region", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import folium

# Load energy data and clean column names
energy_data = pd.read_csv("https://raw.githubusercontent.com/Heleinef/Data-Science-Master_Heleine/refs/heads/main/US_Energy.csv")
energy_data.columns = energy_data.columns.str.replace(r'[.,]', '', regex=True).str.replace(' ', '_')
energy_data['State'] = energy_data['State'].str.upper()  # Ensure state names are in uppercase

# Load U.S. states GeoJSON and prepare for merge
geo_data = gpd.read_file('https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json')
geo_data['name'] = geo_data['name'].str.upper()  # Ensure state names are in uppercase

# Merge the datasets
map_data = geo_data.merge(energy_data, left_on='name', right_on='State')

# Check the result of the merge
print("\nMerged Data Columns and Sample:")
print(map_data.columns.tolist())
print(map_data.head())
```
